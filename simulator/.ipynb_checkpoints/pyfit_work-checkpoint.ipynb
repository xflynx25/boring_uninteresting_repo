```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats

# to do:
# needs to become default/custom uni/bi/multi variate. Maybe nothing else. 
# clean up x-univariate, (add stuff, remove, group better)
# do bivariate, multivariate funcs 
# functions for cleaning, nan checking
# method for data augmentation, particularly how we will deal with categoricals
# once finished, Dataset object is finished, and can work on the Model, feature_engineering, training, and evaluator objects 
# need to have the false positive and false negative rate in the evaluation
class Dataset:
    def __init__(self, data, target, folder='.', sig_figs=-1):
        self.folder = folder
        self.sig_figs = sig_figs
        self.X = self.ensure_dataframe(data)
        self.Y = self.ensure_dataframe(target)
        self.shape_X = self.X.shape
        self.shape_Y = self.Y.shape
        
    
    # the process func will be called in the training and prediction functions
    # although maybe that's a different class, so maybe this just used for the analysis
    def set_process_func(func):
        self.process_func = func

    def ensure_dataframe(self, input_data):
        if not isinstance(input_data, pd.DataFrame):
            if isinstance(input_data, np.ndarray):
                return pd.DataFrame(input_data)   
            elif isinstance(input_data, list):
                return pd.DataFrame(input_data)
            else:
                raise ValueError("Input must be a DataFrame, NumPy array, or a list.")
        return input_data

    def calc_stats(self):
        if self.sig_figs != -1:
            ystats = self.Y.describe().round(self.sig_figs)
            xstats = self.X.describe().round(self.sig_figs)
        else:
            ystats = self.Y.describe()
            xstats = self.X.describe()

        print(f"Y stats:\n{ystats}\n")
        print(f"X stats (first 5 cols):\n{xstats.iloc[:, :5]}\n")

        self.nulls_Y = self.Y.isnull().sum()
        self.nulls_X = self.X.isnull().sum()

        total_nulls_Y = self.nulls_Y.sum()
        total_nulls_X = self.nulls_X.sum()
        
        self.duplicates_X = self.X.duplicated().sum()

        print(f"Total Null counts in Y: {total_nulls_Y}")
        print(f"Columns with Nulls in Y: {self.nulls_Y[self.nulls_Y > 0].index.tolist()}\n")
        
        print(f"Total Null counts in X: {total_nulls_X}")
        print(f"Columns with Nulls in X: {self.nulls_X[self.nulls_X > 0].index.tolist()}\n")

        if total_nulls_Y > 0:
            nulls_Y_indices = self.Y[self.Y.isnull().any(axis=1)].index.tolist()
            print(f"Indices with Nulls in Y: {nulls_Y_indices}\n")
        
        if total_nulls_X > 0:
            null_locations_X = self.X[self.X.isnull().any(axis=1)].stack(dropna=False)
            null_locations_X = null_locations_X[null_locations_X.isna()].index.tolist()
            print(f"First 10 Nulls in X: {null_locations_X[:10]}\n")
            
        print(f"Total Duplicate counts in X: {self.duplicates_X}")
        if self.duplicates_X > 0:
            print(f"Duplicate Rows in X: {self.X[self.X.duplicated()].index.tolist()}\n")

    
    # histogram of Y, boxplot of Y, stats? (outlier_count, mean, var), kernel density plot, cdf
    def visualize_Y(self):
        folder = self.folder + '/plots/y_stats'
        if not os.path.exists(folder):
            os.makedirs(folder)
            
        # Histogram
        plt.figure()
        sns.histplot(self.Y, bins=20, kde=False)
        plt.title('Histogram of Y')
        plt.savefig(folder + '/histogram.png')
        #plt.show()
        
        # Boxplot
        plt.figure()
        sns.boxplot(x=self.Y.squeeze())
        plt.title('Boxplot of Y')
        plt.savefig(folder + '/boxplot.png')
        #plt.show()

        # Stats
        ystats = self.Y.describe()
        print(f"Y stats: {ystats}")

        # Outlier count
        Q1 = self.X.quantile(0.25, numeric_only=True)
        Q3 = self.X.quantile(0.75, numeric_only=True)
        IQR = Q3 - Q1
        outliers = ((self.Y < (Q1 - 1.5 * IQR)) | (self.Y > (Q3 + 1.5 * IQR))).sum()
        print(f"Outlier count: {outliers}")
        
        # Kernel Density Plot
        plt.figure()
        sns.kdeplot(self.Y.iloc[:, 0])  # Assuming Y is a single column, gaussian by default
        plt.title('Kernel Density Plot of Y')
        plt.savefig(folder + '/kde.png')
        #plt.show()
        
        # CDF
        plt.figure()
        sns.ecdfplot(self.Y)
        plt.title('CDF of Y')
        plt.savefig(folder + '/cdf.png')
        #plt.show()
        
        # Combined Plot
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        
        # Histogram
        sns.histplot(self.Y, bins=20, kde=False, ax=axes[0, 0])
        axes[0, 0].set_title('Histogram of Y')

        # Boxplot
        sns.boxplot(x=self.Y.squeeze(), ax=axes[0, 1])
        axes[0, 1].set_title('Boxplot of Y')

        # Kernel Density Plot
        sns.kdeplot(self.Y.iloc[:, 0], ax=axes[1, 0])  # Assuming Y is a single column
        axes[1, 0].set_title('Kernel Density Plot of Y')

        # CDF
        sns.ecdfplot(self.Y, ax=axes[1, 1])
        axes[1, 1].set_title('CDF of Y')

        plt.tight_layout()
        plt.savefig(folder + '/combined_plot.png')
        plt.show()
        
    # for this function we want several things
    # 1) statistics into a csv file, so the stuff from self.describe basically, also if you could add an outlier col
    # 2) quality of fit with gaussian for each variable (also in that csv)
    # 3) the covariance matrix for the X, and also print out the most correlated ones 
    # 4) the correlation of X with Y for all X's, and same here, order this, and then print the top 10 lets say
    # 5) bonus things you think might be useful
    # PCA is probably a good bonus here, unless computatiojnally expensive
    def visualize_X(self, sig_figs=-1):
        float_format = None if sig_figs == -1 else f"%.{sig_figs}g"

        folder = self.folder + '/x_stats'
        if not os.path.exists(folder):
            os.makedirs(folder)

        # 1. Basic Statistics and Outliers
        summary_stats = self.X.describe()

        # Adding outliers based on IQR
        Q1 = self.X.quantile(0.25, numeric_only=True)
        Q3 = self.X.quantile(0.75, numeric_only=True)
        IQR = Q3 - Q1
        outliers_count = ((self.X < (Q1 - 1.5 * IQR)) | (self.X > (Q3 + 1.5 * IQR))).sum()
        summary_stats.loc['outliers'] = outliers_count
        
        # 2. Goodness of Fit (only on numeric columns)
        numeric_cols = self.X.select_dtypes(include=[np.number]).columns

        shapiro_pvals = []
        dropped_cols = 0

        for col in numeric_cols:
            col_data = self.X[col].dropna()
            unique_values = col_data.unique()

            if len(unique_values) > 1:
                p_val = stats.shapiro(col_data)[1]
                shapiro_pvals.append(p_val)
            else:
                print(f"Dropping column {col} for Shapiro test: All values are identical.")
                shapiro_pvals.append(None)  # Or another placeholder value
                dropped_cols += 1

        print(f"Total columns dropped for Shapiro test: {dropped_cols}")
        summary_stats.loc['shapiro_pval'] = pd.Series(shapiro_pvals, index=numeric_cols)
        summary_stats.to_csv(folder + '/summary_stats.csv', float_format=float_format)


        # 3. Covariance Matrix
        cov_matrix = self.X.cov()
        cov_matrix.to_csv(folder + '/cov_matrix.csv', float_format=float_format)

        # Most correlated pairs, maybe can do same thing did with y... show some diff ranges
        corr_matrix = self.X.corr()
        corr_matrix_unstacked = corr_matrix.abs().unstack().sort_values(ascending=False)
        top_correlations = corr_matrix_unstacked[corr_matrix_unstacked < 1.0].head(10)

        print("Top 10 Most Correlated Pairs:")
        print(top_correlations)

        # 4. Correlation with Y
        print('this is YYYYY: ', self.X.shape, self.Y.shape)
        corr_with_y = self.X.corrwith(self.Y.iloc[:,0]).sort_values(ascending=False)
        corr_with_y.to_csv(folder + '/corr_with_y.csv', float_format=float_format)

        # Top 10 Correlations with Y
        print("Top 10 Correlations with Y:")
        print(corr_with_y.head(10))

        # Create buckets for different correlation ranges
        buckets = np.linspace(-1, 1, 21)  # Creates 20 equally spaced bins from -1 to 1
        bucketed_correlations = {}

        # Categorize into buckets and sample 5 from each
        for lower, upper in zip(buckets[:-1], buckets[1:]):
            bucket_key = f"{lower:.1f} to {upper:.1f}"
            in_bucket = corr_with_y[(corr_with_y >= lower) & (corr_with_y < upper)]

            if not in_bucket.empty:
                bucketed_correlations[bucket_key] = in_bucket.sample(min(15, len(in_bucket)))

        # Print sampled correlations in different ranges
        print("Examples of Correlations in Different Ranges:")
        for k, v in bucketed_correlations.items():
            print(f"{k}: {v}")


        # 5. Bonus: Skewness and Kurtosis
        skewness = self.X.skew()
        kurtosis = self.X.kurt()

        summary_stats.loc['skewness'] = skewness
        summary_stats.loc['kurtosis'] = kurtosis

        summary_stats.to_csv(folder + '/summary_stats_with_bonus.csv', float_format=float_format)
        
    
    # heat map correlation matrix
    # scatter plots of top features, with best fit line and residuals
    def default_bivariate_analysis(self):
        pass
    
    # does some more in depth look at a specific column and the relationship with Y
    # i'm thinking for example if i wanted to see things about how the year changed the y-distribution
    def specific_bivariate_analysis(self, col):
        pass
        
    def default_multivariate_analysis(self):
        pass
    
    # try some kernels, or some other things, also look more in depth in the specific 
    # relationship between the cols, not just how they interact with Y
    def specific_multivariate_analysis(self, cols):
        pass
```
